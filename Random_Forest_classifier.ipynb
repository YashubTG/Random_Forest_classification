```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
```

```python
data = load_wine()
```

```python
X = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
Y = pd.DataFrame(data.target)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

```python
rf_model = RandomForestClassifier(n_estimators = 100, max_depth = None, criterion = "gini", random_state = 42, n_jobs = -1)
#n_estimator - no of trees in the random forest algo. standard size = 100, n_jobs - it means that all the cores in the  computer should be used for 
#this algorithm, because this algo takes a lot of computational power, max_depth - max depth a tree can have, criterion - two criteion gini and entropy
#we will use gini here because it taked less computation
```

```python
rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)
y_pred
```

```python
acc = accuracy_score(y_test,y_pred)
```

```python
print(f"accuracy_score{acc*100:.2f}%")
#100%accuracy cause data is small
```

```python
imp = rf_model.feature_importances_
ind = np.argsort(imp)[::-1]
#argssort() - this function gives on the bases of highest value what is the rank of the value
```
